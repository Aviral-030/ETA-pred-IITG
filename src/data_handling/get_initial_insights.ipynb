# get some insights using Pandas

```python
import pandas as pd
import numpy as np
```

```python
# import our data
delivery_data = pd.read_csv("../../dataset/raw_data.csv")
print(delivery_data)
```

```python
# slight overview
print("train vs test data: ")
print(delivery_data['data'].value_counts(normalize=False), "\n")

print("list of all the hubs in India")
print(pd.concat([delivery_data["destination_name"], delivery_data["source_name"]]).unique(), "\n")

```

```python
print("list of tracks where OSRM time was close to actual time")
delivery_data["actual_time"] = pd.to_numeric(delivery_data["actual_time"])
delivery_data["osrm_time"] = pd.to_numeric(delivery_data["osrm_time"])
filterred_data = delivery_data.loc[delivery_data["osrm_time"] < 1.25*delivery_data["actual_time"], ["source_name", "destination_name"]].drop_duplicates()
print(filterred_data)
```
